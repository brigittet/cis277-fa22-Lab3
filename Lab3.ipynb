### 8. What is the average rating for each genre (Action, Adventure, Comedy, Crime, Family, Fantasy, Mystery, Sci-Fi, and Thriller)?

```python
action_games = df[df.Action == True]
print('Action: ', action_games.rating.mean())

adv_games = df[df.Adventure == True]
print('Adventure: ', adv_games.rating.mean())

com_games = df[df.Comedy == True]
print('Comedy: ', com_games.rating.mean())

crime_games = df[df.Crime == True]
print('Crime: ', crime_games.rating.mean())

fam_games = df[df.Family == True]
print('Family: ', fam_games.rating.mean())

fant_games = df[df.Fantasy == True]
print('Fantasy: ', fant_games.rating.mean())

myst_games = df[df.Mystery == True]
print('Mystery: ', myst_games.rating.mean())

scifi_games = df[df['Sci-Fi'] == True]
print('Sci-Fi: ', scifi_games.rating.mean())

thril_games = df[df.Thriller == True]
print('Thriller: ', thril_games.rating.mean())
```

### 12. What is the most common certificate of games with ratings greater than 8.0 and votes greater than 1000?

### 13. For games with ratings greater than the 75th percentile for ratings and a number of votes greater than the 75th percentile for votes, what is the most common genre?

```python
rating_threshold = df.rating.quantile(0.75)
votes_threshold = df.votes.quantile(0.75)

good_games = df[(df.rating > rating_threshold) & (df.votes > votes_threshold)]

genres = {'Action': 0, 'Adventure': 0, 'Comedy': 0, 'Crime': 0, 'Family': 0, 'Fantasy': 0, 'Mystery': 0, 'Sci-Fi': 0, 'Thriller': 0}

for genre in genres:
    genres[genre] = len(good_games[good_games[genre] == True])

most_common = max(genres.values())
for genre in genres:
    if genres[genre] == most_common:
        print(genre, most_common)
```
